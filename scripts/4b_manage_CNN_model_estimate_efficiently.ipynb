```python
import openpyxl

channels_num_used = set()
for which_channel in which_channels_list:
    channels_num_used.add(len(which_channel))

filename_sg_part = ""
for channel_num_used in channels_num_used:
    if channel_num_used == 0:
        filename_sg_part += "all_"
    else:
        filename_sg_part += f"{channel_num_used}_"
filename_sg_part += "SG"

combined_results.to_excel(rf"{setting_yaml_file_base}_Flatten{cnn_model_user.should_use_flatten_data}_{filename_sg_part}_result.xlsx", index=False, sheet_name=setting_yaml_file_base)
# combined_results.to_excel(r"240108_wo_all_1_result.xlsx", index=False)
```
